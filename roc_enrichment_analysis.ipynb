```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, precision_recall_curve
import seaborn as sns
from pathlib import Path

# ============================================================================
# CONFIGURATION
# ============================================================================

# Input files
ACTIVITY_CSV = "P08173_random_subset.csv"        # File with compound activities/labels
DOCKING_RESULTS_CSV = "boltz_analysis_results.csv"  # File with docking results

# Column names in activity CSV
COMPOUND_ID_COL = "ligand_chembl_id"             # Column with compound identifiers (e.g., "CHEMBL192")
ACTION_TYPE_COL = "action_type"                  # Column with activity labels

# Labels for classification (case-insensitive matching)
POSITIVE_CLASS_LABELS = ["AGONIST", "agonist"]   # Labels considered as positive class
NEGATIVE_CLASS_LABELS = ["ANTAGONIST", "antagonist"]  # Labels considered as negative class

# Column names in docking results CSV
DOCKING_COMPOUND_ID_COL = "molecule_name"        # Column with compound identifiers
SCORING_METRICS = [                              # Metrics to evaluate (higher = better binding)
    "affinity_probability_binary",
    "affinity_pred_value",
    "confidence_score",
    "iptm",
    "ligand_iptm"
]
DISTANCE_METRICS = [                             # Distance metrics (lower = better binding)
    "ligand_to_binding_site_distance"
]

# Enrichment plot settings
ENRICHMENT_FRACTIONS = [0.01, 0.05, 0.10, 0.20]  # Calculate enrichment at 1%, 5%, 10%, 20%

# Output settings
OUTPUT_DIR = "roc_enrichment_plots"              # Directory to save plots
FIGURE_DPI = 300                                 # Resolution for saved figures

# ============================================================================
# FUNCTIONS
# ============================================================================

def load_and_merge_data(activity_csv, docking_csv, 
                        compound_id_col, action_type_col,
                        docking_compound_id_col,
                        positive_labels, negative_labels):
    """
    Load activity and docking data, merge them, and create binary labels.
    
    Returns:
        pd.DataFrame: Merged dataframe with binary labels
    """
    # Load activity data
    activity_df = pd.read_csv(activity_csv)
    print(f"Loaded {len(activity_df)} records from {activity_csv}")
    
    # Load docking results
    docking_df = pd.read_csv(docking_csv)
    print(f"Loaded {len(docking_df)} docking results from {docking_csv}")
    
    # Create a clean activity dataframe with unique compounds
    # Keep only rows with positive or negative labels
    activity_clean = activity_df[[compound_id_col, action_type_col]].copy()
    activity_clean[action_type_col] = activity_clean[action_type_col].str.upper().str.strip()
    
    # Filter for compounds with known labels
    positive_labels_upper = [label.upper() for label in positive_labels]
    negative_labels_upper = [label.upper() for label in negative_labels]
    
    activity_clean = activity_clean[
        activity_clean[action_type_col].isin(positive_labels_upper + negative_labels_upper)
    ]
    
    # Remove duplicates, keeping first occurrence
    activity_clean = activity_clean.drop_duplicates(subset=[compound_id_col], keep='first')
    
    # Create binary label (1 = positive class, 0 = negative class)
    activity_clean['label'] = activity_clean[action_type_col].apply(
        lambda x: 1 if x in positive_labels_upper else 0
    )
    
    print(f"\nActivity summary:")
    print(f"  Positive class ({positive_labels}): {(activity_clean['label'] == 1).sum()}")
    print(f"  Negative class ({negative_labels}): {(activity_clean['label'] == 0).sum()}")
    
    # Merge with docking results
    merged_df = docking_df.merge(
        activity_clean[[compound_id_col, 'label', action_type_col]],
        left_on=docking_compound_id_col,
        right_on=compound_id_col,
        how='inner'
    )
    
    print(f"\nAfter merging: {len(merged_df)} compounds with both docking scores and labels")
    print(f"  Positive class: {(merged_df['label'] == 1).sum()}")
    print(f"  Negative class: {(merged_df['label'] == 0).sum()}")
    
    return merged_df

def calculate_enrichment_factor(y_true, y_scores, fraction):
    """
    Calculate enrichment factor at a given fraction of the database.
    
    Args:
        y_true: True binary labels
        y_scores: Predicted scores (higher = better)
        fraction: Fraction of database to consider (e.g., 0.01 for top 1%)
    
    Returns:
        float: Enrichment factor
    """
    n_total = len(y_true)
    n_positives = np.sum(y_true)
    n_selected = max(1, int(fraction * n_total))
    
    # Sort by scores (descending)
    sorted_indices = np.argsort(y_scores)[::-1]
    top_n = sorted_indices[:n_selected]
    
    # Count positives in top N
    n_positives_found = np.sum(y_true[top_n])
    
    # Calculate enrichment factor
    # EF = (n_positives_found / n_selected) / (n_positives / n_total)
    random_rate = n_positives / n_total
    actual_rate = n_positives_found / n_selected
    
    if random_rate > 0:
        ef = actual_rate / random_rate
    else:
        ef = 0
    
    return ef, n_positives_found, n_selected

def plot_roc_curve(y_true, y_scores, metric_name, output_dir, 
                   positive_label="Agonist", negative_label="Antagonist"):
    """
    Plot ROC curve for a given metric.
    """
    # Calculate ROC curve
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    
    # Create figure
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Plot ROC curve
    ax.plot(fpr, tpr, color='darkorange', lw=2, 
            label=f'ROC curve (AUC = {roc_auc:.3f})')
    
    # Plot random classifier line
    ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', 
            label='Random classifier (AUC = 0.500)')
    
    # Styling
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate', fontsize=12)
    ax.set_ylabel('True Positive Rate', fontsize=12)
    ax.set_title(f'ROC Curve: {metric_name}\n{positive_label} vs {negative_label}', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc="lower right", fontsize=11)
    ax.grid(True, alpha=0.3)
    
    # Add text box with interpretation
    interpretation = "Excellent" if roc_auc >= 0.9 else \
                    "Good" if roc_auc >= 0.8 else \
                    "Fair" if roc_auc >= 0.7 else \
                    "Poor" if roc_auc >= 0.6 else "Random"
    
    textstr = f'Performance: {interpretation}'
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(0.6, 0.1, textstr, transform=ax.transAxes, fontsize=11,
            verticalalignment='top', bbox=props)
    
    plt.tight_layout()
    
    # Save figure
    output_path = Path(output_dir) / f"roc_{metric_name.replace(' ', '_')}.png"
    plt.savefig(output_path, dpi=FIGURE_DPI, bbox_inches='tight')
    print(f"Saved ROC curve to {output_path}")
    
    plt.show()
    
    return roc_auc, fpr, tpr

def plot_enrichment_curve(y_true, y_scores, metric_name, output_dir,
                         enrichment_fractions,
                         positive_label="Agonist", negative_label="Antagonist"):
    """
    Plot enrichment curve showing cumulative recall of positives.
    """
    n_total = len(y_true)
    n_positives = np.sum(y_true)
    
    # Sort by scores (descending)
    sorted_indices = np.argsort(y_scores)[::-1]
    y_true_sorted = y_true[sorted_indices]
    
    # Calculate cumulative recall
    fractions = np.linspace(0, 1, n_total)
    cumulative_positives = np.cumsum(y_true_sorted)
    recall = cumulative_positives / n_positives
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot enrichment curve
    ax.plot(fractions * 100, recall * 100, color='blue', lw=2.5, 
            label=f'{metric_name}')
    
    # Plot random selection line
    ax.plot([0, 100], [0, 100], color='gray', lw=2, linestyle='--', 
            label='Random selection')
    
    # Plot ideal curve (all positives at top)
    ideal_x = [0, (n_positives / n_total) * 100, 100]
    ideal_y = [0, 100, 100]
    ax.plot(ideal_x, ideal_y, color='green', lw=2, linestyle=':', 
            label='Ideal enrichment')
    
    # Calculate and display enrichment factors
    ef_text = []
    for frac in enrichment_fractions:
        ef, n_found, n_selected = calculate_enrichment_factor(y_true, y_scores, frac)
        ef_text.append(f"EF{int(frac*100)}%: {ef:.2f} ({n_found}/{n_selected})")
        
        # Add vertical line at this fraction
        ax.axvline(x=frac*100, color='red', linestyle=':', alpha=0.3, lw=1)
    
    # Styling
    ax.set_xlim([0, 100])
    ax.set_ylim([0, 105])
    ax.set_xlabel('% of Database Screened', fontsize=12)
    ax.set_ylabel(f'% of {positive_label}s Found', fontsize=12)
    ax.set_title(f'Enrichment Plot: {metric_name}\n{positive_label} vs {negative_label}', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc="lower right", fontsize=11)
    ax.grid(True, alpha=0.3)
    
    # Add enrichment factors text box
    textstr = '\n'.join(ef_text)
    props = dict(boxstyle='round', facecolor='lightyellow', alpha=0.8)
    ax.text(0.55, 0.25, textstr, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', bbox=props, family='monospace')
    
    plt.tight_layout()
    
    # Save figure
    output_path = Path(output_dir) / f"enrichment_{metric_name.replace(' ', '_')}.png"
    plt.savefig(output_path, dpi=FIGURE_DPI, bbox_inches='tight')
    print(f"Saved enrichment plot to {output_path}")
    
    plt.show()
    
    return ef_text

def plot_precision_recall_curve(y_true, y_scores, metric_name, output_dir,
                                positive_label="Agonist", negative_label="Antagonist"):
    """
    Plot Precision-Recall curve.
    """
    precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    
    # Calculate baseline (random classifier)
    baseline = np.sum(y_true) / len(y_true)
    
    # Create figure
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Plot PR curve
    ax.plot(recall, precision, color='blue', lw=2, 
            label=f'PR curve (AUC = {pr_auc:.3f})')
    
    # Plot baseline
    ax.axhline(y=baseline, color='gray', lw=2, linestyle='--',
              label=f'Random classifier (baseline = {baseline:.3f})')
    
    # Styling
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('Recall', fontsize=12)
    ax.set_ylabel('Precision', fontsize=12)
    ax.set_title(f'Precision-Recall Curve: {metric_name}\n{positive_label} vs {negative_label}', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc="upper right", fontsize=11)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # Save figure
    output_path = Path(output_dir) / f"pr_curve_{metric_name.replace(' ', '_')}.png"
    plt.savefig(output_path, dpi=FIGURE_DPI, bbox_inches='tight')
    print(f"Saved PR curve to {output_path}")
    
    plt.show()
    
    return pr_auc

def plot_score_distributions(merged_df, metric_name, output_dir,
                            positive_label="Agonist", negative_label="Antagonist"):
    """
    Plot distribution of scores for positive and negative classes.
    """
    fig, axes = plt.subplots(2, 1, figsize=(10, 10))
    
    # Get data for each class
    positive_scores = merged_df[merged_df['label'] == 1][metric_name].dropna()
    negative_scores = merged_df[merged_df['label'] == 0][metric_name].dropna()
    
    # Plot 1: Overlapping histograms
    ax = axes[0]
    ax.hist(positive_scores, bins=30, alpha=0.6, label=positive_label, 
            color='green', edgecolor='black')
    ax.hist(negative_scores, bins=30, alpha=0.6, label=negative_label, 
            color='red', edgecolor='black')
    ax.set_xlabel(metric_name, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title(f'Score Distribution: {metric_name}', fontsize=13, fontweight='bold')
    ax.legend(fontsize=11)
    ax.grid(True, alpha=0.3, axis='y')
    
    # Add statistics
    stats_text = f"{positive_label}: ={positive_scores.mean():.3f}, ={positive_scores.std():.3f}\n"
    stats_text += f"{negative_label}: ={negative_scores.mean():.3f}, ={negative_scores.std():.3f}"
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    # Plot 2: Box plot
    ax = axes[1]
    data_to_plot = [positive_scores, negative_scores]
    bp = ax.boxplot(data_to_plot, labels=[positive_label, negative_label],
                    patch_artist=True, widths=0.6)
    
    # Color box plots
    colors = ['lightgreen', 'lightcoral']
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
    
    ax.set_ylabel(metric_name, fontsize=12)
    ax.set_title(f'Box Plot Comparison: {metric_name}', fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y')
    
    # Perform statistical test
    from scipy.stats import mannwhitneyu
    statistic, p_value = mannwhitneyu(positive_scores, negative_scores, alternative='two-sided')
    
    sig_text = f"Mann-Whitney U test\np-value = {p_value:.4e}\n"
    if p_value < 0.001:
        sig_text += "*** (p < 0.001)"
    elif p_value < 0.01:
        sig_text += "** (p < 0.01)"
    elif p_value < 0.05:
        sig_text += "* (p < 0.05)"
    else:
        sig_text += "n.s. (not significant)"
    
    ax.text(0.98, 0.98, sig_text, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))
    
    plt.tight_layout()
    
    # Save figure
    output_path = Path(output_dir) / f"distributions_{metric_name.replace(' ', '_')}.png"
    plt.savefig(output_path, dpi=FIGURE_DPI, bbox_inches='tight')
    print(f"Saved distribution plots to {output_path}")
    
    plt.show()

def create_summary_table(results_dict, output_dir):
    """
    Create a summary table of all metrics' performance.
    """
    summary_data = []
    
    for metric, results in results_dict.items():
        row = {
            'Metric': metric,
            'ROC AUC': results['roc_auc'],
            'PR AUC': results['pr_auc']
        }
        
        # Add enrichment factors
        for ef_str in results['enrichment_factors']:
            # Parse "EF1%: 2.5 (5/20)" format
            ef_label = ef_str.split(':')[0]
            ef_value = float(ef_str.split(':')[1].split('(')[0].strip())
            row[ef_label] = ef_value
        
        summary_data.append(row)
    
    summary_df = pd.DataFrame(summary_data)
    summary_df = summary_df.sort_values('ROC AUC', ascending=False)
    
    # Save to CSV
    output_path = Path(output_dir) / "performance_summary.csv"
    summary_df.to_csv(output_path, index=False)
    print(f"\nSaved performance summary to {output_path}")
    
    # Display table
    print("\n" + "="*80)
    print("PERFORMANCE SUMMARY")
    print("="*80)
    print(summary_df.to_string(index=False))
    print("="*80)
    
    return summary_df

# ============================================================================
# MAIN ANALYSIS
# ============================================================================

def main():
    """
    Main analysis pipeline.
    """
    # Create output directory
    Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
    
    print("="*80)
    print("ROC AND ENRICHMENT ANALYSIS")
    print("="*80)
    
    # Load and merge data
    print("\n1. Loading and merging data...")
    merged_df = load_and_merge_data(
        ACTIVITY_CSV, 
        DOCKING_RESULTS_CSV,
        COMPOUND_ID_COL,
        ACTION_TYPE_COL,
        DOCKING_COMPOUND_ID_COL,
        POSITIVE_CLASS_LABELS,
        NEGATIVE_CLASS_LABELS
    )
    
    # Display ligand class counts
    n_agonists = (merged_df['label'] == 1).sum()
    n_antagonists = (merged_df['label'] == 0).sum()
    total_ligands = len(merged_df)
    
    print("\n" + "="*80)
    print("LIGAND CLASS SUMMARY")
    print("="*80)
    print(f"Total ligands with docking results: {total_ligands}")
    print(f"  Agonists:     {n_agonists} ({n_agonists/total_ligands*100:.1f}%)")
    print(f"  Antagonists:  {n_antagonists} ({n_antagonists/total_ligands*100:.1f}%)")
    print("="*80)
    
    # Store results for summary
    results_dict = {}
    
    # Analyze scoring metrics (higher is better)
    print("\n2. Analyzing scoring metrics (higher = better binding)...")
    for metric in SCORING_METRICS:
        if metric not in merged_df.columns:
            print(f"\nWarning: Metric '{metric}' not found in docking results. Skipping.")
            continue
        
        # Remove NaN values
        valid_data = merged_df[['label', metric]].dropna()
        
        if len(valid_data) < 10:
            print(f"\nWarning: Not enough valid data for metric '{metric}'. Skipping.")
            continue
        
        print(f"\n{'='*80}")
        print(f"Analyzing: {metric}")
        print(f"{'='*80}")
        
        y_true = valid_data['label'].values
        y_scores = valid_data[metric].values
        
        # ROC curve
        roc_auc, fpr, tpr = plot_roc_curve(
            y_true, y_scores, metric, OUTPUT_DIR,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Enrichment plot
        ef_text = plot_enrichment_curve(
            y_true, y_scores, metric, OUTPUT_DIR, ENRICHMENT_FRACTIONS,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Precision-Recall curve
        pr_auc = plot_precision_recall_curve(
            y_true, y_scores, metric, OUTPUT_DIR,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Distribution plots
        plot_score_distributions(
            valid_data, metric, OUTPUT_DIR,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Store results
        results_dict[metric] = {
            'roc_auc': roc_auc,
            'pr_auc': pr_auc,
            'enrichment_factors': ef_text
        }
    
    # Analyze distance metrics (lower is better)
    print("\n3. Analyzing distance metrics (lower = better binding)...")
    for metric in DISTANCE_METRICS:
        if metric not in merged_df.columns:
            print(f"\nWarning: Metric '{metric}' not found in docking results. Skipping.")
            continue
        
        # Remove NaN values
        valid_data = merged_df[['label', metric]].dropna()
        
        if len(valid_data) < 10:
            print(f"\nWarning: Not enough valid data for metric '{metric}'. Skipping.")
            continue
        
        print(f"\n{'='*80}")
        print(f"Analyzing: {metric}")
        print(f"{'='*80}")
        
        y_true = valid_data['label'].values
        # Invert distance scores (multiply by -1) so higher is better
        y_scores = -valid_data[metric].values
        
        # ROC curve
        roc_auc, fpr, tpr = plot_roc_curve(
            y_true, y_scores, metric, OUTPUT_DIR,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Enrichment plot
        ef_text = plot_enrichment_curve(
            y_true, y_scores, metric, OUTPUT_DIR, ENRICHMENT_FRACTIONS,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Precision-Recall curve
        pr_auc = plot_precision_recall_curve(
            y_true, y_scores, metric, OUTPUT_DIR,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Distribution plots (use original distance values)
        plot_score_distributions(
            valid_data, metric, OUTPUT_DIR,
            POSITIVE_CLASS_LABELS[0], NEGATIVE_CLASS_LABELS[0]
        )
        
        # Store results
        results_dict[metric] = {
            'roc_auc': roc_auc,
            'pr_auc': pr_auc,
            'enrichment_factors': ef_text
        }
    
    # Create summary table
    print("\n4. Creating performance summary...")
    summary_df = create_summary_table(results_dict, OUTPUT_DIR)
    
    print(f"\n{'='*80}")
    print("ANALYSIS COMPLETE!")
    print(f"All plots and summary saved to: {OUTPUT_DIR}")
    print(f"{'='*80}")

if __name__ == "__main__":
    main()
```
