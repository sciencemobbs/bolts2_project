```python
#!/usr/bin/env python3
#importing necessary libraries
import sys
import sqlite3
import pandas as pd
from tqdm.auto import tqdm
from molbloom import buy
import os
import argparse
from typing import Optional
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Descriptors, AllChem, Draw
from rdkit.Chem.MolStandardize import rdMolStandardize
from sklearn.manifold import TSNE
import numpy as np

#location of the database
DEFAULT_DB_FILENAME = "/home/jmobbs/src/adv_bootcamp/week2/chembl_36/chembl_36_sqlite/chembl_36.db"

#default location of the mappings
DEFAULT_MAPPING_FILENAME = "/home/jmobbs/src/adv_bootcamp/week2/chembl_uniprot_mapping.txt"

#Enter uniprot ID here
UNIPROT_ID = "P08173"  # Change this to your desired UniProt ID

# Chemical property filters (adjust these values as needed)
MW_MIN = 200
MW_MAX = 500
ALOGP_MIN = -1
ALOGP_MAX = 5

# Random subset size
RANDOM_SUBSET_SIZE = 100  # Number of molecules to randomly select for detailed analysis

def get_query() -> str:
    """Returns the SQL query to fetch assay data for a given ChEMBL ID."""
    sql = """SELECT
        d.doc_id,
        d.doi,
        d.title,
        a.assay_id,
        a.description,
        cr.compound_key,
        cs.molregno,
        cs.canonical_smiles,
        m.chembl_id AS ligand_chembl_id,
        act.standard_type,
        act.standard_value,
        act.standard_relation,
        act.standard_units,
        act.pchembl_value,
        act.activity_comment,
        dm.action_type
    FROM
        target_dictionary td
            JOIN
        assays a ON td.tid = a.tid
            JOIN
        docs d ON a.doc_id = d.doc_id
            JOIN
        activities act ON a.assay_id = act.assay_id AND d.doc_id = act.doc_id
            JOIN
        compound_records cr ON act.record_id = cr.record_id
            JOIN
        compound_structures cs ON cr.molregno = cs.molregno
            LEFT JOIN
        molecule_dictionary m ON cs.molregno = m.molregno
            LEFT JOIN
        drug_mechanism dm ON cs.molregno = dm.molregno
    WHERE
        td.chembl_id = ?;"""
    return sql

def get_chembl_id(uniprot_id: str, mapping_filename: str) -> Optional[str]:
    """
    Retrieves the ChEMBL ID for a given UniProt ID from the mapping file.
    """
    if not os.path.exists(mapping_filename):
        print(f"Error: Can't find mapping file {mapping_filename}", file=sys.stderr)
        return None

    uniprot_df = pd.read_csv(mapping_filename, sep="	", comment='#',
                             names=['uniprot_id', 'chembl_id', 'target_name', 'target_type'])
    
    uniprot_res = uniprot_df.query("uniprot_id == @uniprot_id and target_type == 'SINGLE PROTEIN'")
    
    if len(uniprot_res) == 0:
        print(f"No ChEMBL ID found for UniProt {uniprot_id}")
        return None
        
    return uniprot_res.chembl_id.values[0]

def standardize_molecules(df: pd.DataFrame) -> pd.DataFrame:
    """
    Standardize molecules by removing salts and generating standardized SMILES.
    Also removes duplicate molecules based on standardized SMILES.
    Consolidates action_type information when removing duplicates.
    """
    print("\nStandardizing molecules (removing salts)...")
    
    # Initialize the fragment remover and uncharger
    fragment_remover = rdMolStandardize.LargestFragmentChooser()
    uncharger = rdMolStandardize.Uncharger()
    
    initial_count = len(df)
    std_smiles_list = []
    valid_indices = []
    
    for idx, smiles in enumerate(tqdm(df['canonical_smiles'], desc="Standardizing")):
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            try:
                # Remove salts (keep largest fragment)
                mol = fragment_remover.choose(mol)
                # Neutralize charges
                mol = uncharger.uncharge(mol)
                # Generate standardized SMILES
                std_smiles = Chem.MolToSmiles(mol)
                std_smiles_list.append(std_smiles)
                valid_indices.append(idx)
            except:
                # If standardization fails, skip this molecule
                continue
        
    # Create new dataframe with only valid molecules
    df = df.iloc[valid_indices].copy()
    df['std_smiles'] = std_smiles_list
    
    invalid_count = initial_count - len(df)
    if invalid_count > 0:
        print(f"Removed {invalid_count} molecules that could not be standardized")
    
    # Remove duplicates based on standardized SMILES
    print("\nRemoving duplicate molecules...")
    before_dedup = len(df)
    
    # Consolidate action_type before removing duplicates
    # Keep all unique action types for each molecule
    if 'action_type' in df.columns:
        df['action_type'] = df.groupby('std_smiles')['action_type'].transform(
            lambda x: '; '.join(sorted(set(str(i) for i in x.dropna().unique()))) if x.notna().any() else None
        )
    
    df = df.drop_duplicates(subset=['std_smiles'], keep='first')
    after_dedup = len(df)
    duplicates_removed = before_dedup - after_dedup
    
    print(f"Removed {duplicates_removed} duplicate molecules ({duplicates_removed/before_dedup*100:.1f}%)")
    print(f"Unique molecules remaining: {after_dedup}")
    
    return df

def calculate_molecular_properties(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate molecular weight and ALogP for each compound using RDKit.
    """
    print("\nCalculating molecular properties...")
    
    mol_weights = []
    alogps = []
    
    for smiles in tqdm(df['canonical_smiles'], desc="Computing properties"):
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            mol_weights.append(Descriptors.MolWt(mol))
            alogps.append(Descriptors.MolLogP(mol))
        else:
            mol_weights.append(None)
            alogps.append(None)
    
    df['molecular_weight'] = mol_weights
    df['alogp'] = alogps
    
    # Remove rows where properties couldn't be calculated
    initial_count = len(df)
    df = df.dropna(subset=['molecular_weight', 'alogp'])
    removed = initial_count - len(df)
    
    if removed > 0:
        print(f"Removed {removed} compounds with invalid structures")
    
    return df

def filter_by_properties(df: pd.DataFrame, mw_min: float, mw_max: float, 
                         logp_min: float, logp_max: float) -> pd.DataFrame:
    """
    Filter compounds by molecular weight and ALogP.
    """
    print(f"\n{'='*60}")
    print("STEP 1: Filtering by Chemical Properties")
    print(f"{'='*60}")
    
    initial_count = len(df)
    
    print(f"Initial compound count: {initial_count}")
    print(f"\nApplying filters:")
    print(f"  Molecular Weight: {mw_min} - {mw_max}")
    print(f"  ALogP: {logp_min} - {logp_max}")
    
    # Apply filters
    filtered_df = df[
        (df['molecular_weight'] >= mw_min) &
        (df['molecular_weight'] <= mw_max) &
        (df['alogp'] >= logp_min) &
        (df['alogp'] <= logp_max)
    ].copy()
    
    final_count = len(filtered_df)
    removed = initial_count - final_count
    
    print(f"\nResults:")
    print(f"  Compounds passing filters: {final_count}")
    print(f"  Compounds removed: {removed}")
    
    if final_count > 0:
        print(f"\nProperty ranges in filtered set:")
        print(f"  MW: {filtered_df['molecular_weight'].min():.1f} - {filtered_df['molecular_weight'].max():.1f}")
        print(f"  ALogP: {filtered_df['alogp'].min():.2f} - {filtered_df['alogp'].max():.2f}")
    
    return filtered_df

def filter_by_purchasability(df: pd.DataFrame) -> pd.DataFrame:
    """
    Filter compounds by purchasability using MolBloom.
    """
    print(f"\n{'='*60}")
    print("STEP 2: Filtering by Purchasability")
    print(f"{'='*60}")
    
    initial_count = len(df)
    print(f"Checking purchasability for {initial_count} compounds...")
    
    # Check purchasability
    df['purchasable'] = [buy(smi, canonicalize=True) for smi in tqdm(df['canonical_smiles'], desc="Checking")]
    
    # Filter to only purchasable compounds
    filtered_df = df[df['purchasable'] == True].copy()
    
    final_count = len(filtered_df)
    percentage = (final_count / initial_count * 100) if initial_count > 0 else 0
    
    print(f"\nResults:")
    print(f"  Purchasable compounds: {final_count} ({percentage:.1f}%)")
    print(f"  Non-purchasable compounds: {initial_count - final_count}")
    
    return filtered_df

def get_random_subset(df: pd.DataFrame, n: int = 100) -> pd.DataFrame:
    """
    Select a random subset of molecules.
    
    Parameters:
    df: DataFrame of filtered molecules
    n: Number of molecules to sample (default 100)
    """
    print(f"\n{'='*60}")
    print("STEP 3: Random Subset Selection")
    print(f"{'='*60}")
    
    if len(df) < n:
        print(f"⚠️  Warning: Only {len(df)} molecules available.")
        n = len(df)
        print(f"Sampling all {n} molecules.")
    else:
        print(f"Selecting {n} random molecules from {len(df)} available")
    
    random_subset = df.sample(n=n, random_state=42)
    
    print(f"\nSubset summary statistics:")
    print(random_subset[['molecular_weight', 'alogp']].describe())
    
    return random_subset

def generate_fingerprints(df: pd.DataFrame) -> np.ndarray:
    """
    Generate Morgan fingerprints for molecules.
    
    Parameters:
    df: DataFrame with std_smiles column
    
    Returns:
    Array of binary fingerprints
    """
    print(f"\n{'='*60}")
    print("STEP 4: Fingerprint Generation")
    print(f"{'='*60}")
    
    fingerprints = []
    valid_indices = []
    
    for idx, smiles in enumerate(tqdm(df['std_smiles'], desc="Generating fingerprints")):
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            # Generate Morgan fingerprint (radius=2, 2048 bits)
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
            fingerprints.append(np.array(fp))
            valid_indices.append(idx)
    
    fingerprints = np.array(fingerprints)
    
    print(f"Generated {len(fingerprints)} fingerprints (2048 bits each)")
    
    return fingerprints, df.iloc[valid_indices].reset_index(drop=True)

def run_tsne_analysis(fingerprints: np.ndarray, output_file: str = 'tsne_plot.png'):
    """
    Run t-SNE analysis on molecular fingerprints and plot results.
    
    Parameters:
    fingerprints: Array of binary fingerprints
    output_file: Output filename for plot
    """
    print(f"\n{'='*60}")
    print("STEP 5: t-SNE Analysis")
    print(f"{'='*60}")
    
    if len(fingerprints) < 2:
        print("⚠️  Not enough molecules for t-SNE analysis")
        return None
    
    # Adjust perplexity based on sample size
    # perplexity should be less than n_samples
    perplexity = min(30, len(fingerprints) - 1)
    
    print(f"Running t-SNE with perplexity={perplexity}...")
    print("(This may take a moment...)")
    
    tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity, max_iter=1000)
    coords = tsne.fit_transform(fingerprints)
    
    # Create plot
    plt.figure(figsize=(10, 8))
    plt.scatter(coords[:, 0], coords[:, 1], alpha=0.6, s=50, c='blue', edgecolors='black')
    plt.title('t-SNE Analysis of Molecular Fingerprints\n(Random Subset Chemical Space Distribution)', 
              fontsize=14, fontweight='bold')
    plt.xlabel('t-SNE Dimension 1', fontsize=12)
    plt.ylabel('t-SNE Dimension 2', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    # Add text annotation
    plt.text(0.02, 0.98, f'n = {len(fingerprints)} molecules', 
             transform=plt.gca().transAxes, fontsize=10,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.savefig(output_file, dpi=300, bbox_inches='tight')
    print(f"t-SNE plot saved to: {output_file}")
    plt.show()
    
    return coords

def display_random_molecules(df: pd.DataFrame, n: int = 10, output_file: str = 'random_molecules.png'):
    """
    Display chemical structures of random molecules from the subset.
    
    Parameters:
    df: DataFrame with molecular data
    n: Number of molecules to display (default 10)
    output_file: Output filename for the image
    """
    print(f"\n{'='*60}")
    print("STEP 6: Chemical Structure Visualization")
    print(f"{'='*60}")
    
    sample_size = min(n, len(df))
    sample = df.sample(n=sample_size, random_state=42)
    
    mols = []
    legends = []
    
    print(f"Displaying {sample_size} random molecules from the subset:\n")
    
    for idx, (i, row) in enumerate(sample.iterrows(), 1):
        mol = Chem.MolFromSmiles(row['std_smiles'])
        if mol:
            mols.append(mol)
            legend = f"MW: {row['molecular_weight']:.1f}\nALogP: {row['alogp']:.2f}"
            legends.append(legend)
            
            # Print SMILES
            print(f"{idx}. SMILES: {row['std_smiles']}")
            print(f"   MW: {row['molecular_weight']:.1f}, ALogP: {row['alogp']:.2f}")
            if 'molregno' in row:
                print(f"   ChEMBL molregno: {row['molregno']}")
            print()
    
    if mols:
        # Create grid image
        img = Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(300, 300), 
                                   legends=legends, returnPNG=False)
        img.save(output_file)
        print(f"Chemical structures saved to: {output_file}")
    else:
        print("⚠️  No valid molecules to display")
    
    return mols

def process_uniprot_id(uniprot_id: str, db_filename: str, mapping_filename: str,
                       mw_min: float, mw_max: float, logp_min: float, logp_max: float,
                       random_subset_size: int = 100):
    """
    Processes a single UniProt ID to fetch and save ChEMBL assay data.
    """
    if not os.path.exists(db_filename):
        print(f"Error: Can't find ChEMBL database file {db_filename}", file=sys.stderr)
        return

    chembl_id = get_chembl_id(uniprot_id, mapping_filename)
    if not chembl_id:
        return

    print(f"\n{'='*60}")
    print(f"Searching for ChEMBL ID {chembl_id} (from UniProt {uniprot_id})")
    print(f"{'='*60}")

    try:
        conn = sqlite3.connect(db_filename)
        sql = get_query()
        df = pd.read_sql_query(sql, conn, params=(chembl_id,))
        conn.close()
    except sqlite3.Error as e:
        print(f"Database error: {e}", file=sys.stderr)
        return

    print(f"\nFound {len(df)} total records from database")
    
    # Show action_type distribution if available
    if 'action_type' in df.columns:
        action_counts = df['action_type'].value_counts(dropna=False)
        print(f"\nAction type distribution:")
        for action, count in action_counts.head(10).items():
            if pd.isna(action):
                print(f"  Unknown/None: {count}")
            else:
                print(f"  {action}: {count}")

    if df.empty:
        print("No data found. Exiting.")
        return

    # Standardize molecules and remove duplicates early
    df = standardize_molecules(df)
    
    if df.empty:
        print("No valid molecules after standardization. Exiting.")
        return

    # Calculate molecular properties using RDKit
    df = calculate_molecular_properties(df)
    
    if df.empty:
        print("No valid compounds after property calculation. Exiting.")
        return

    # Filter by chemical properties
    df = filter_by_properties(df, mw_min, mw_max, logp_min, logp_max)
    
    if df.empty:
        print("\nNo compounds passed property filters. Exiting.")
        return

    # Filter by purchasability
    df = filter_by_purchasability(df)
    
    if df.empty:
        print("\nNo purchasable compounds found. Exiting.")
        return

    # Get random subset of molecules
    random_subset = get_random_subset(df, n=random_subset_size)
    
    # Write random subset to CSV
    subset_output_filename = f"{uniprot_id}_random_subset.csv"
    random_subset.to_csv(subset_output_filename, index=False)
    print(f"\nRandom subset saved to: {subset_output_filename}")
    
    # Generate fingerprints for the random subset
    fingerprints, random_subset = generate_fingerprints(random_subset)
    
    if len(fingerprints) == 0:
        print("\nNo valid fingerprints generated. Exiting.")
        return
    
    # Run t-SNE analysis
    tsne_output_filename = f"{uniprot_id}_tsne_plot.png"
    run_tsne_analysis(fingerprints, output_file=tsne_output_filename)
    
    # Display 10 random molecules from the subset
    molecules_output_filename = f"{uniprot_id}_random_molecules.png"
    display_random_molecules(random_subset, n=10, output_file=molecules_output_filename)

    # Write the full filtered results to a CSV file (all molecules before random sampling)
    output_filename = f"{uniprot_id}_filtered.csv"
    df.to_csv(output_filename, index=False)
    
    print(f"\n{'='*60}")
    print("FINAL RESULTS")
    print(f"{'='*60}")
    print(f"Total filtered compounds: {len(df)}")
    print(f"Random subset size: {len(random_subset)}")
    print(f"\nOutput files created:")
    print(f"  1. {output_filename} - All filtered compounds")
    print(f"  2. {subset_output_filename} - Random subset ({random_subset_size} molecules)")
    print(f"  3. {tsne_output_filename} - t-SNE visualization")
    print(f"  4. {molecules_output_filename} - 10 random molecule structures")
    print(f"{'='*60}\n")

def main():
    """
    Main function to run the data fetching process.
    """
    # Use the hardcoded UniProt ID and property filters
    process_uniprot_id(UNIPROT_ID, DEFAULT_DB_FILENAME, DEFAULT_MAPPING_FILENAME,
                       MW_MIN, MW_MAX, ALOGP_MIN, ALOGP_MAX, RANDOM_SUBSET_SIZE)

if __name__ == "__main__":
    main()
```
